```python
#!pip install auto-sklearn
import pandas as pd
import sklearn
import autosklearn
import autosklearn.classification
from benchmark_utils import timer
```

```python
X_train = pd.read_csv("../datasets/dont-overfit/train.csv")
X_test = pd.read_csv("../datasets/dont-overfit/test.csv")

## Change target dtype 
y_train = pd.DataFrame(X_train.target,dtype='int')
# Drop id and target from train frame
X_train.drop(columns=['target','id'],inplace=True)
## Copy test id's and drop id from test frame
id_test = X_test.id
X_test.drop(columns=['id'],inplace=True)
```

```python
X_train.head()
```

## Treinamento

```python
start_time = timer(None)
ak = autosklearn.classification.AutoSklearnClassifier(ml_memory_limit=129024)

ak.fit(X_train, y_train,metric=autosklearn.metrics.accuracy)

time = timer(start_time)
```

```python
results_df = ak.cv_results_
results_df
```

```python
preds =  ak.predict(X_test)
print(preds)

```

```python
print(time)
```

```python

preds_train = ak.predict(X_train)

print("Accuracy score", sklearn.metrics.accuracy_score(preds_train, y_train))
```

```python
submission = pd.DataFrame({
        "id": id_test,
        "target": preds
    })

submission.to_csv("../submit_files/dont_overfit_autosk_submission.csv", index=False)
```
