```python
#!pip install h2o
import pandas as pd
import h2o
from h2o.automl import H2OAutoML
from benchmark_utils import timer
```

```python
h2o.init()
X_train = h2o.import_file('../datasets/dont-overfit/train.csv',header=1)
X_test = h2o.import_file('../datasets/dont-overfit/test.csv',header=1)
```

```python
X_train.head()
```

## Treinamento

```python
start_time = timer(None)

x = X_train.columns
y = "target"

X_train[y] = X_train[y].asfactor()

# (limited to 1 hour max runtime by default)
aml = H2OAutoML()
aml.train(x=x, y=y, training_frame=X_train)
time = timer(start_time)

```

```python
print(time)
```

```python
# View the AutoML Leaderboard
lb = aml.leaderboard
lb.head(rows=lb.nrows)  # Print all rows instead of default (10 rows)
```

```python
# The leader model is stored here
aml.leader
```

```python
preds = aml.leader.predict(X_test)
print(preds)

```

```python
h2o.save_model(aml.leader, path = "Dont-overfit_h2o")
```

```python
from sklearn import metrics

X_train_df = pd.read_csv("../datasets/dont-overfit/train.csv")
## Change target dtype
y_train = pd.DataFrame(X_train_df.target,dtype='int')

preds_train = aml.leader.predict(X_train)

print("Auc", metrics.roc_auc_score(preds_train['predict'].as_data_frame(), y_train))
```
