```python
#!pip install auto-sklearn
import pandas as pd
import autosklearn.classification
from benchmark_utils import timer
```

```python
X_train = pd.read_csv("../datasets/porto-seguro/train.csv")
X_test = pd.read_csv("../datasets/porto-seguro/test.csv")

y_train = pd.to_numeric(X_train['target'])
# Drop id and target from train frame
X_train.drop(columns=['target','id'],inplace=True)
## Copy test id's and drop id from test frame
id_test = X_test.id
X_test.drop(columns=['id'],inplace=True)
```

```python
X_train.head()
```

## Treinamento

```python
start_time = timer(None)
ak = autosklearn.classification.AutoSklearnClassifier(ml_memory_limit=129024)

ak.fit(X_train, y_train)

time = timer(start_time)
```

```python

preds =  ak.predict(X_test)


print(preds)
```

```python
print(time)
```

```python
submission = pd.DataFrame({
        "id": id_test,
        "target": preds
    })

submission.to_csv("../submit_files/porto_seguro_autosk_submission.csv", index=False)
```
