```python
#!pip install tpot
import pandas as pd
from tpot import TPOTClassifier
from benchmark_utils import timer
```

```python
X_train = pd.read_csv("../datasets/porto-seguro/train.csv")
X_test = pd.read_csv("../datasets/porto-seguro/test.csv")

## Change target dtype 
y_train = pd.DataFrame(X_train.target,dtype='int')
# Drop id and target from train frame
X_train.drop(columns=['target','id'],inplace=True)
## Copy test id's and drop id from test frame
id_test = X_test.id
X_test.drop(columns=['id'],inplace=True)
```

```python
X_train.head()
```

## Treinamento

```python
start_time = timer(None)
#tp = TPOTClassifier(generations=50, population_size=25, random_state=42, verbosity=3)

tp = TPOTClassifier(generations=5, population_size=50, random_state=42, verbosity=2)

tp.fit(X_train, y_train)
time = timer(start_time)
```

```python
tp.export('tpot_pipeline_porto-seguro.py')
print(time)
preds = tp.predict(X_test)
print(preds)
```

```python
submission = pd.DataFrame({
        "id": id_test,
        "target": preds
    })

submission.to_csv("../submit_files/porto-seguro_tpot_submission.csv", index=False)
```
