```python
#!pip install auto-sklearn
import pandas as pd
import autosklearn
import autosklearn.regression
from benchmark_utils import timer
```

```python
X_train = pd.read_csv("../datasets/taxi-fare/train.csv")
X_test = pd.read_csv("../datasets/taxi-fare/test.csv")

```

```python
X_train.info()
```

```python

## Change target dtype 
y_train = pd.to_numeric(X_train['fare_amount'])
# Drop id and target from train frame
X_train.drop(columns=['fare_amount','key'],inplace=True)
## Copy test id's and drop id from test frame
id_test = X_test.key
X_test.drop(columns=['key'],inplace=True)
#X_train = pd.get_dummies(X_train)
#X_test = pd.get_dummies(X_test)
```

```python
X_train.head()
```

```python
X_train.info()
```

## Treinamento

```python
start_time = timer(None)
ak = autosklearn.regression.AutoSklearnRegressor(ml_memory_limit=129024)

ak.fit(X_train, y_train)

time = timer(start_time)
```

```python
results_df = ak.cv_results_
print(results_df)
preds =  ak.predict(X_test)


print(preds)
```

```python
print(time)
```

```python
submission = pd.DataFrame({
        "id": id_test,
        "target": preds
    })

submission.to_csv("../submit_files/taxi_fare_autosk_submission.csv", index=False)
```
